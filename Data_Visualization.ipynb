```python
df=tips[['total_bill','tip']]
df.head()
```

```python
stats.spearmanr(df.total_bill,df.tip)
```

```python
stats.pearsonr(df.total_bill,df.tip)
```

```python
df.corr(method='spearman')
```

```python
df.corr(method='pearson')
```
