## Data Cleaning
 In our dataset all 16 columns have equal number of datapoints, 4024 entries. So, data cleaning is not required here.

## Data Encoding

```python
data1['A Stage']
data1['Estrogen Status']
data1['Progesterone Status']
data1['Status']
```

```python
data1['T Stage ']=data1['T Stage '].map({'T1':1,'T2':2,'T3':3,'T4':4})
data1['N Stage']=data1['N Stage'].map({'N1':1,'N2':2,'N3':3})
data1['6th Stage']=data1['6th Stage'].map({'IIA':1,'IIIA':2,'IIB':3,'IIIB':4,'IIIC':5})
data1['differentiate']=data1['differentiate'].map({'Undifferentiated':1,'Poorly differentiated':2,'Moderately differentiated':3,'Well differentiated':4})
data1['Grade']=data1['Grade'].map({'1':1,'2':2,'3':3,' anaplastic; Grade IV':4})
```

```python
data1.info()
```

```python
data1['A Stage']=data1['A Stage'].map({'Regional':1,'Distant':0})
data1['Estrogen Status']=data1['Estrogen Status'].map({'Positive':1,'Negative':0})
data1['Progesterone Status']=data1['Progesterone Status'].map({'Positive':1,'Negative':0})
data1['Status']=data1['Status'].map({'Alive':1,'Dead':0})
```

```python
data1.info()
```

```python
data2=get_dummies(data1,['Race','Marital Status'],dtype=int)
```

```python
data2.info()
```

```python
data2.to_csv('Breast Cancer Data_Atik.csv')
```

## Defining X and Y

```python
y=data2['Status']
x=data2.drop('Status',axis=1)
print (x.shape)
print (y.shape)
```

## Scaling X

```python
from sklearn.preprocessing import StandardScaler
x_scaled=StandardScaler().fit_transform(x)
DataFrame(x_scaled)
```

## Data Splitting

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.25, random_state=67)
print (x_train.shape)
print (x_test.shape)
print (y_train.shape)
print (y_test.shape)
```

## Data Balancing (for Training set)

```python
y_train.value_counts()
```

```python
from imblearn.over_sampling import SMOTE
x_train,y_train=SMOTE(random_state=103).fit_resample(x_train,y_train)
```
