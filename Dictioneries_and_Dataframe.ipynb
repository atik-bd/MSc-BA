```python
#1. Write a Pandas program to append a list of dictioneries or series to an existing DataFrame and display the combined data.
import pandas as pd
# Create the initial DataFrame with student details
data = {'student_id': ['S1', 'S2', 'S3', 'S4', 'S5'],
        'name': ['Danniella Fenton', 'Ryder Storey', 'Bryce Jensen', 'Ed Bernal', 'Kwame Morin'],
        'marks': [200, 210, 190, 222, 199]}
student_detail = pd.DataFrame(data)

# Display the original DataFrame
print("Original DataFrame:\n", student_detail)

# Create a dictionary for the new student
new_entry = {'student_id': ['S6'],'name': ['Scarlette Fisher'],'marks': [205]}

new_student=pd.DataFrame(new_entry)
# Append the new student to the DataFrame

student_detail=pd.concat([student_detail, new_student], ignore_index=True)
# Display the combined DataFrame
print("\nCombined DataFrame:")
print(student_detail)
```

```python
#2. Write a Pandas program to join the two given dataframes along rows and merge with another dataframe along the common column id.
import pandas as pd
data1={'student_id':['S1','S2','S3','S4','S5'],'name':['Danniella Fenton','Ryder Storey','Bryce Jensen','Ed Bernal','Kwame Morin'],
       'marks':[200,210,190,222,199]}
student_detail1=pd.DataFrame(data1)
print("First DataFrame:\n",student_detail1)

data2={'student_id':['S4','S5','S6','S7','S8'],'name':['Scarlette Fisher','Carla Williamson','Dante Morse','Kaiser William','Madeeha Preston'],
       'marks':[201,200,198,219,201]}
student_detail2=pd.DataFrame(data2)
print("\nSecond DataFrame:\n",student_detail2)

combined_data=pd.concat([student_detail1,student_detail2],ignore_index=True)
print("\nCombined Dataframe:\n",combined_data)

data3= {'student_id': ['S1', 'S2', 'S3', 'S4', 'S5', 'S7', 'S8', 'S9', 'S10', 'S11', 'S12', 'S13'],
        'exam_id': [23, 45, 12, 67, 21, 55, 33, 14, 56, 83, 88, 12]}
exam_data=pd.DataFrame(data3)
print("\nThird DataFrame:\n",exam_data)

merged_data=pd.merge(combined_data,exam_data,on='student_id')
print("\nMerged Dataframe:\n",merged_data)
```

```python
#3. Write a Pandas program to join the two dataframes using the common column of both dataframes.
import pandas as pd
data1={'student_id':['S1','S2','S3','S4','S5'],'name':['Danniella Fenton','Ryder Storey','Bryce Jensen','Ed Bernal','Kwame Morin'],
       'marks':[200,210,190,222,199]}
student_data1=pd.DataFrame(data1)
print("First DataFrame:\n",student_data1)

data2={'student_id':['S4','S5','S6','S7','S8'],'name':['Scarlette Fisher','Carla Williamson','Dante Morse','Kaiser William','Madeeha Preston'],
       'marks':[201,200,198,219,201]}
student_data2=pd.DataFrame(data2)
print("\nSecond DataFrame:\n",student_data2)

result=pd.merge(student_data1,student_data2,on='student_id')
print("\nMerged Data(inner join):\n",result)
```

```python
# 4. Write a Pandas program to join the two dataframes with matching records from both sides where available.
import pandas as pd
data1={'student_id':['S1','S2','S3','S4','S5'],'name':['Danniella Fenton','Ryder Storey','Bryce Jensen','Ed Bernal','Kwame Morin'],
       'marks':[200,210,190,222,199]}
student_data1=pd.DataFrame(data1)
print("First DataFrame:\n",student_data1)

data2={'student_id':['S4','S5','S6','S7','S8'],'name':['Scarlette Fisher','Carla Williamson','Dante Morse','Kaiser William','Madeeha Preston'],
       'marks':[201,200,198,219,201]}
student_data2=pd.DataFrame(data2)
print("\nSecond DataFrame:\n",student_data2)

result=pd.merge(student_data1,student_data2,on='student_id',how='outer')
print("\nJoined Data(outer join):\n",result)
```

```python
# 5. Write a Pandas program to join (left join) the two dataframes using keys from left dataframe only.
import pandas as pd
data1={'key1':['K0','K0','K1','K2'],'key2':['K0','K1','K0','K1'],'P':['P0','P1','P2','P3'],'Q':['Q0','Q1','Q2','Q3']}
df1=pd.DataFrame(data1)
print("First DataFrame:\n",df1)

data2={'key1':['K0','K1','K1','K2'],'key2':['K0','K0','K0','K0'],'R':['R0','R1','R2','R3'],'S':['S0','S1','S2','S3']}
df2=pd.DataFrame(data2)
print("\nSecond DataFrame:\n",df2)

result=pd.merge(df1,df2,how='left',on=['key1','key2'])
print("\nResult:\n",result)
```

```python
# 6. Write a Pandas program to create a line plot of the historical stock prices of Alphabet Inc. between two specific dates.
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv('/content/drive/MyDrive/DataSets/alphabet_stock_data.csv')
# Convert the 'Date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Set the date as the index
data.set_index('Date', inplace=True)

# Define the date range for filtering
start_date = input("Input Start Date (YYYY-MM-DD): ")
end_date = input("Input End Date (YYYY-MM-DD): ")

# Filter the data between the specified dates
filtered_data = data.loc[start_date:end_date]

# Plot the closing prices
plt.figure(figsize=(12, 6))
plt.plot(filtered_data['Close'], label='Closing Price', color='blue')
plt.title('Historical Stock Prices of Alphabet Inc. (GOOGLE)')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# 7. Write a Pandas program to create a line plot of the opening, closing stock prices of Alphabet Inc. between two specific dates.
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('/content/drive/MyDrive/DataSets/alphabet_stock_data.csv')

# Convert the 'Date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Set the date as the index
data.set_index('Date', inplace=True)

# Define the date range for filtering
start_date = input("Input Start Date (YYYY-MM-DD): ")
end_date = input("Input End Date (YYYY-MM-DD): ")

# Filter the data between the specified dates
filtered_data = data.loc[start_date:end_date]

# Plot the opening and closing prices
plt.figure(figsize=(12, 6))
plt.plot(filtered_data.index,filtered_data['Open'], label='Opening Price', color='blue')
plt.plot(filtered_data.index,filtered_data['Close'], label='Closing Price', color='red')
plt.title('Alphabet Inc. Stock Prices (Opening & Closing)')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# 10. Write a Pandas program to create a stacked bar plot of opening, closing stock prices of Alphabet Inc. between two specific dates.
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('/content/drive/MyDrive/DataSets/alphabet_stock_data.csv')

data['Date']=pd.to_datetime(data['Date'])
data.set_index('Date',inplace=True)
start_date = input("Input Start Date (YYYY-MM-DD): ")
end_date = input("Input End Date (YYYY-MM-DD): ")
filtered_data = data.loc[start_date:end_date]
plt.figure(figsize=(12,6))
plt.bar(filtered_data.index,filtered_data['Open'],label='Opening Price',color='blue')
plt.bar(filtered_data.index, filtered_data['Close'] - filtered_data['Open'], bottom=filtered_data['Open'], label='Closing Price', color='red')
plt.title('Alphabet Inc. Stock Prices (Opening & Closing)')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```
