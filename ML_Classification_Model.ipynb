## Data Encoding

```python
# Binary Encoding_ converting categorical features into numerical binary values
data1['default']=data1['default'].map({'no':1,'yes':0})
data1['housing']=data1['housing'].map({'no':1,'yes':0})
data1['loan']=data1['loan'].map({'no':1,'yes':0})
data1['y']=data1['y'].map({'no':1,'yes':0})
data1.info()
```

```python
# Label encoding for ordinal categorical variables
data1['month']=data1['month'].map({'jan':1,'feb':2,'mar':3,'apr':4,'may':5,'jun':6,'jul':7,'aug':8,'sep':9,'oct':10,'nov':11,'dec':12})
```

```python
# One-Hot Encoding_ converts categorical features into multiple binary columns
data2 = get_dummies(data1,columns=['job','marital','education','contact','poutcome'],dtype=int)
data2.info()
```

## Dividing the dataset

```python
X=data2.drop('y',axis=1)
Y=data2['y']
print(X.shape)
print(Y.shape)
```

## Data Scaling

```python
from sklearn.preprocessing import StandardScaler
X_scaled=StandardScaler().fit_transform(X)
DataFrame(X_scaled)
```

## Data Splitting

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X_scaled,Y,test_size=0.2,random_state=80)
print("X_train:", X_train.shape)
print("X_test:",X_test.shape)
print("Y_train:",Y_train.shape)
print("Y_test:",Y_test.shape)
```

## Data Balancing

```python
Y_train.value_counts()
```

```python
from imblearn.over_sampling import SMOTE
X_train,Y_train=SMOTE(random_state=70).fit_resample(X_train,Y_train)
Y_train.value_counts()
```

**bold text**# random forest Using pipeline

# Random Forest Classifier

```python
# Random Forest Classifier using pipeline (method 3)
from imblearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn import ensemble
RF_classifier = Pipeline([('balancing', SMOTE(random_state = 25)),('classification', ensemble.RandomForestClassifier(criterion='entropy', max_features='sqrt', random_state=100))]) # building classifier
no_trees = {'classification__n_estimators': [250,255,260,265,270,275,280]}
grid_search1 = GridSearchCV(estimator=RF_classifier, param_grid=no_trees, scoring='recall', cv=5)
grid_search1.fit(X_scaled, Y)
best_parameters = grid_search1.best_params_
print(best_parameters)
best_result = grid_search1.best_score_
print(best_result)
```

```python
# Random Forest Classifier (method 1)
from sklearn import ensemble
RF_classifier1 =ensemble.RandomForestClassifier(n_estimators=270, criterion='entropy', max_features='sqrt', random_state=50)  # building model
RF_classifier1.fit(X_train,Y_train)#training
Y_predRF=RF_classifier1.predict(X_test)# testing
```

```python
# Evaluation RF
# Accuracy and confusion matrix
from sklearn import metrics
Accuracy=metrics.accuracy_score(Y_test, Y_predRF) # calculating accuaracy
print("Random Forest Classifier\nAccuracy: ", Accuracy) # Is this a good metric??
con_matrix = metrics.confusion_matrix(Y_test, Y_predRF)
print ("confusion_matrix: \n", con_matrix)
recall = metrics.recall_score(Y_test, Y_predRF)
print ("recall: ",recall)
percision=metrics.precision_score(Y_test, Y_predRF)
print("percision: ",percision)
f1=metrics.f1_score(Y_test, Y_predRF)
print('f1= ',f1)
```

# Support Vector Classifier (SVC)

```python
# Support Vector Classifier using pipeline (method 3)
from imblearn.pipeline import Pipeline
from sklearn import svm
from sklearn.model_selection import GridSearchCV
SV_classifier2 = Pipeline([('balancing', SMOTE(random_state = 1)),('classification', svm.SVC() ) ]) # bulilding classifier
kernels_c = {'classification__kernel': ['linear','poly','rbf','sigmoid'], 'classification__C': [.01,.1,1,10,100]}
grid_search1 = GridSearchCV(estimator=SV_classifier2, param_grid=kernels_c, scoring='recall', cv=5)
grid_search1.fit(X_scaled, Y)
best_parameters = grid_search1.best_params_
print(best_parameters)
best_result = grid_search1.best_score_
print(best_result)
```

```python
#building SVM using method#1
from sklearn.svm import SVC
SV_classifier1 = SVC(kernel='poly', C=0.01)  # building classifier
SV_classifier1.fit(X_train, Y_train) # training
Y_predSVC= SV_classifier1.predict(X_test)  #testing
```

```python
# Evaluation SVC
# Accuracy and confusion matrix
from sklearn import metrics
Accuracy=metrics.accuracy_score(Y_test, Y_predSVC) # calculating accuaracy
print("Support Vector Classifier\nAccuracy: ", Accuracy) # Is this a good metric??
con_matrix = metrics.confusion_matrix(Y_test, Y_predSVC)
print ("confusion_matrix: \n", con_matrix)
recall = metrics.recall_score(Y_test, Y_predSVC)
print ("recall: ",recall)
percision=metrics.precision_score(Y_test, Y_predSVC)
print("percision: ",percision)
f1=metrics.f1_score(Y_test, Y_predSVC)
print('f1= ',f1)
```
